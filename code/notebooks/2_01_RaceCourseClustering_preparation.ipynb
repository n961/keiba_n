```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
pd.set_option('display.max_columns', 100)
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Hiragino Maru Gothic Pro', 'Yu Gothic', 'Meirio', 'Takao', 'IPAexGothic', 'IPAPGothic', 'VL PGothic', 'Noto Sans CJK JP']
```

```python
df_race_result = pd.read_csv('race_result_inter.csv')
df_race_info = pd.read_csv('race_info_inter.csv')
df_horse_data = pd.read_csv('horse_data.csv')
df_race_data = pd.merge(df_race_result, df_race_info, on='race_id')
df_race_data
```

```python
df_top_three_result = df_race_data[df_race_data['着順'] <= 3]
df_top_three_result_under_open = df_top_three_result[df_top_three_result['race_class_prize'] != 'オープン']
df_top_three_result_open = df_top_three_result[df_top_three_result['race_class_prize'] == 'オープン']
```

```python
course_name_list = set(df_race_data['course_name'])
list_df_course = {}
for course_name in course_name_list:
    list_df_course[course_name] = (df_top_three_result_under_open[df_top_three_result_under_open['course_name'] == course_name])
    
list_df_course_open = {}
for course_name in course_name_list:
    list_df_course_open[course_name] = (df_top_three_result_open[df_top_three_result_open['course_name'] == course_name])
```

```python
index = 1
course_count = len(course_name_list)
fig = plt.figure(figsize=(10,50),dpi=200)
fig.suptitle('各競馬場における、着順3着以内馬の最終コーナー通過順位', y=0.9)
fig.text(0.125, 0.89, 'オープン未満', color='blue')
fig.text(0.55, 0.89, 'オープン以上', color='blue')
for key in list_df_course:
    ax1 = fig.add_subplot(course_count, 2, index)
    ax2 = fig.add_subplot(course_count, 2, index + 1)
    ax1.hist(list_df_course[key]['passing4'], bins=list_df_course[key]['passing4'].nunique())
    ax2.hist(list_df_course_open[key]['passing4'], bins=list_df_course_open[key]['passing4'].nunique())
    ax1.set_title(key)
    ax2.set_title(key)
    index = index + 2
```
